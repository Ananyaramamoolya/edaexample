```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dirty-employee.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Full_Name']=df['Full_Name'].fillna("").astype(str).str.strip().str.title()
```

```python
df["Age"]=df["Age"].fillna(0)
df_Filtered=df[(df["Age"]>=1)&(df["Age"]<=100)]
```

```python
df["Joining_Date"]=pd.to_datetime(df["Joining_Date"],errors="coerce")
df["Joining_Date"]=df["Joining_Date"].dt.strftime("%Y-%m-%d")
```

```python
df["Phone"]=df["Phone"].astype(str).str.replace(r"[^0-9]","",regex=True)
```

```python
df["Marks"]=pd.to_numeric(df["Marks"],errors="coerce")
df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.to_csv("Cleaned_employee_data.csv",index=False)
```
